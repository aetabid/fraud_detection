```python
trans_features = trans_df.groupby("account_id").agg(
    total_tx=("trans_id", "count"),
    total_amount=("amount", "sum"),
    avg_amount=("amount", "mean"),
    avg_balance=("balance", "mean")
).reset_index()
```

```python
main_df = main_df.merge(trans_features, on="account_id")
```

```python
main_df
```

```python
main_df = main_df.merge(account_df, on="account_id")
```

```python
main_df = main_df.merge(trans_features, on="account_id", how="left")
```
